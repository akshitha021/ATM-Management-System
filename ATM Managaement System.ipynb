```python
#console based ATM
#users dictionary
users = {
    "67893453": ['u_1', 50000, 1234],
    "34567765": ['u_2', 15000, 7843],
    "23456756": ['u_3', 68000, 7823],
    "65423455": ['u_4', 180000, 8933]  }
print("***Welcome to Atm***")
print()
ac_no = input("Please enter ur acc no: ")
transactions = []
if ac_no in users:
    max_attempts = 3
    attempt = 1
    active_session = True
    while attempt <= max_attempts:
        user_pin = int(input("please enter your pin: "))

        if user_pin == users[ac_no][-1]:
            print("Pin matches proceed")
            break
        else:
            rem_attempts = (max_attempts - attempt)

            if rem_attempts == 0:
                print("Account blocked please visit after 24hrs.")
                active_session = False
                break

            print(f"entered wrong pin & have only {rem_attempts} chances.")
            attempt += 1

    while True and active_session:
        print("Choose any option. Please enter option number.")
        print("1. Deposits")
        print("2. Withdraw")
        print("3. Balance enquiry")
        print("4. Mini Statement")
        print("5. Pin change")
        print("6. Exit")
        choice = int(input("Please enter option num: "))

        if choice == 1:
            print("You have to choose deposit")
            amount = int(input("Enter the amount you want to deposit: "))
            if amount > 0:
                users[ac_no][1] += amount
                transactions.append(f"{amount} is deposited.")
                print("Amount deposited.")
            else: 
                print("Invalid amount")

        elif choice == 2:
            amount = int(input("Enter the amount you want to withdraw: "))
            if amount > users[ac_no][1]:
                print("Insufficient Funds")
            else:
                users[ac_no][1] -= amount
                transactions.append(f"Rs. {amount} withdrawn")
                print("Please collect cash")

        elif choice == 3:
            print("You have chosen for balance enquiry.")
            print(f"Your current balance is: Rs.{users[ac_no[1]]}")

        elif choice == 4:
            if len(transactions) <= 5:
                for i in transactions[::-1]:
                    print(i)
            if len(transactions) > 5:
                n = len(transactions)
                for i in transactions[n-5:: -1]:
                    print(i)

        elif choice == 5:
            user_pin = int(input("Enter your existing pin: "))
            if users[ac_no][-1] == user_pin:
                print("Please proceed with pin change..")
                new_pin = int(input("Please enter new pin:"))
                confirm_pin = int(input("Please confirm new pin: "))
                if new_pin == confirm_pin:
                    users[ac_no][-1] = new_pin
                    print()
                else:
                    print("Pin confirmation failed .")
            else:
                print("Incorrect pin. Please try again.")

        elif choice == 6:
            print("Thankyou visit again!!!")
else:
    print("Acc doesn't exit thankyou!")
```
